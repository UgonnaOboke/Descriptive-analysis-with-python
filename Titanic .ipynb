```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.set_style('whitegrid')
```

```python
#1 plot a joint plot of fare against age

sns.jointplot(x = 'fare', y = "age", data = titanic)
```

```python
#2 distribution plot of fare

sns.distplot(titanic['fare'], kde = False, bins = 30,
            color = 'black')
```

```python
#3 boxplot 

sns.boxplot(x='class', y = 'age', data = titanic, palette ='rainbow')
```

```python
#4 swarmplot

sns.swarmplot(x='class', y = 'age', data = titanic, 
              palette ='Set2')
```

```python
#5 Countplot


sns.countplot(x = 'sex', data = titanic)
```

```python
# 6 Correleation heatmap plot

tc = titanic.corr()
sns.heatmap(tc, annot = True, cmap ='coolwarm')
```

```python
#7 Facetgrid plot of age vs sex

fc = sns.FacetGrid(data = titanic, col = 'sex')
fc.map(sns.distplot, 'age')
```
